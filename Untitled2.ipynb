```python
import pandas as pd

df = pd.read_csv('data/stats.csv', encoding='utf-8')
df.columns = ['Time', 'Mode', 'Nick', 'Hero', 'Stat', 'Value']
df['Time'] = pd.to_datetime(df['Time']).dt.date
df.set_index('Time')
```

```python
slug = df[(df['Nick'] == 'slug') & (df['Mode'] == 'Quickplay') & (df['Hero'] == 'ALL HEROES') & (df['Stat'].isin(['Games Won','Games Lost'])) ].pivot(index='Time', columns='Stat', values='Value')
slug['Games Lost'] = slug['Games Lost'].diff()
slug['Games Won'] = slug['Games Won'].diff()
```

```python
print(slug)
```

```python
print(slug['Games Won'].index[3])
```

```python
ax = slug.plot.barh(stacked=True, width=0.7, figsize=(8, len(slug['Games Won'])/2.7))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
ax.grid(axis='x')
ax.text(2, 6, "TEST")
```
